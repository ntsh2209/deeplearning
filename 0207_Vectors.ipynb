```python
def plot_vectors(vecs):
  colors = ['r', 'b', 'g', 'y']
  i = 0
  for vec in vecs:
    plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles='xy', scale=1, color=colors[i%len(colors)])
    i += 1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```python
plot_vectors([(0,0,3,4), (0,0,-3,4), (0,0,-3,-2), (0,0,4,-1)])
```

```python
a = np.asarray([5, 4])
b = np.asarray([-3, -2])
```

$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y$   

```python
a_dot_b = np.dot(a, b)
```

```python
print(a_dot_b)
```

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```python
a_b = np.dot(a, b)/np.linalg.norm(b)
```

```python
print(a_b)
```

$\vec{a_b} = a_b \hat{b} = a_b \frac{\vec{b}}{|\vec{b}|}$

```python
vec_a_b = (a_b/np.linalg.norm(b))*b
print(vec_a_b)
```

# Linear combination

$\vec{c} = w_1 \vec{a} + w_2 \vec{b}$

```python
def plot_linear_combination(a, b, w1, w2):
  
  plt.quiver(0,0,a[0],a[1], scale_units='xy', angles='xy', scale=1, color='r')
  plt.quiver(0,0,b[0],b[1], scale_units='xy', angles='xy', scale=1, color='b')
  
  c = w1 * a + w2 * b
  
  plt.quiver(0,0,c[0],c[1], scale_units='xy', angles='xy', scale=1, color='g')

  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```python
a = np.asarray([3, 4])
b = np.asarray([1.5, 2])
```

```python
plot_linear_combination(a, b, -1, 1)
```

```python
def plot_span(a, b):
  
  for i in range(1000):
    w1 = (np.random.random(1) - 0.5) * 3
    w2 = (np.random.random(1) - 0.5) * 3
    c = w1 * a + w2 * b
    plt.quiver(0,0,c[0],c[1], scale_units='xy', angles='xy', scale=1, color='g')
    
  plt.quiver(0,0,a[0],a[1], scale_units='xy', angles='xy', scale=1, color='r')
  plt.quiver(0,0,b[0],b[1], scale_units='xy', angles='xy', scale=1, color='b')

  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

## MP Neuron Class

```python
class MPNeuron:
    def __init__(self):
        self.b = None
    
    def model(self,x):
        return (sum(x) >= self.b)
    
    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)
    
    # learning algorithm
    def fit(self, X, Y):
        accuracy= {}

        for b in range(X.shape[1]+1): # shape 1 is number of columns
            self.b = b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y_pred,Y)

        best_b = max(accuracy, key=accuracy.get) # key=accuracy.get returns the value of the dictionary with the highest accuracy
        self.b = best_b
        print("Best b:", self.b, "Accuracy:", accuracy[best_b])
        
```

```python
mp_neuron = MPNeuron()
mp_neuron.fit(X_binarised_train, Y_train)
```

## Perceptron Class

$y = 1, if \sum_i w_i x_i >= b$




```python
class Perceptron:
    def __init__(self):
        self.b = None
        self.w = None
        
    
    def model(self,x):
        return 1 if np.dot(self.w, x) >= self.b else 0
        

    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)
    
    def fit(self, X, Y,epochs=10, lr=1): # lr is learning rate, not used in this simple, this is also hyper parameter
        # learning algorithm
        self.w = np.ones(X.shape[1])  # initialize weights to 1
        self.b = 0                     # initialize bias to 0
        accuracy = {}
        # epochs is hyper parameter
        max_accuracy = 0
        # checkpointing
        check_w = self.w
        check_b = self.b
        wt_matrix = []

        for epoch in range(epochs):
            # print("Epoch:", epoch)
            for x,y in zip(X,Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.w = self.w + x*lr
                    self.b = self.b + 1*lr
                elif y == 0 and y_pred == 1:
                    self.w = self.w - x*lr
                    self.b = self.b - 1*lr
            accuracy[epoch] = accuracy_score(self.predict(X), Y)
            if accuracy[epoch] > max_accuracy:
                max_accuracy = accuracy[epoch]
                # save the model
                check_w = self.w
                check_b = self.b
            wt_matrix.append(self.w)
        print("Max accuracy:", max_accuracy)
        self.w = check_w
        self.b = check_b
        
        plt.plot(accuracy.values())
        plt.ylim(0,1)
        plt.show()
        best_epoch = max(accuracy, key=accuracy.get)
        print("Best epoch:", best_epoch, "Accuracy:", accuracy[best_epoch])
        return np.array(wt_matrix)
        
    
```
