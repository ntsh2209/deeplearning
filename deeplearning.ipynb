# vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.quiver(0,0,3,4)
```

```python
plt.quiver(0,0,3,4, angles='xy', scale_units='xy', scale=1,color='r')
plt.quiver(0,0,-3,4, angles='xy', scale_units='xy', scale=1,color='g')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

```python
def plot_vectors(vectors):
    colors = ['r', 'g', 'b', 'c', 'm', 'y']
    for vector in vectors:
        plt.quiver(vector[0],vector[1], vector[2], vector[3], angles='xy', scale_units='xy', scale=1,color=colors[random.randint(0, len(colors)-1)])
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid()
    plt.show()
```

```python
plot_vectors([(0, 0, 3, 4), (0, 0, -3, 4), (1, 1, 2, 2)])  # Example usage of the plot_vectors function
```

```python
vecs = [np.asarray([0, 0, 3, 4]), np.asarray([0, 0, -3, 4]), np.asarray([1, 1, 2, 2])]
```

```python
plot_vectors((vecs[0],vecs[1]))
```

```python
vecs[0]+vecs[1]  # Adding two vectors
```

```python
plot_vectors([vecs[0]+vecs[1]-vecs[2]])  # Example usage of the plot_vectors function
```

```python
a = np.asarray([3,4])
b= np.asarray([4,-1])
```

\[
\vec{a} \cdot \vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y
\]

```python
a_dot_b = np.dot(a, b)  # Dot product of vectors a and b
```

```python
a_dot_b
```

```python
# projection =
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```python
a_b
```

```python
vec_a_b = a_b/np.linalg.norm(b) * b  # Projection of vector a onto vector b
```

```python
vec_a_b
```
